 Write a Python program to convert only the characters at odd indexes to lowercase. Example: Input: PYTHON Output: PyThOn

```python
text=input("enter the inpyt:")
s=""
for i in range(len(text)):
    if i%2==0:
        s=s+text[i].lower()
    else:
        s=s+text[i]
print(s)
```
