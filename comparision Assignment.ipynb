7. Set of unique squares (1–10).

8. Unique vowels from a string.
Input: "comprehension"
Output:
{'o', 'e', 'i'}

```python
def unique_vowels(s):
    vowels = "aeiou"
    result = set()

    for char in s:
        if char in vowels:
            result.add(char)

    return result


print(unique_vowels("comprehension"))

    
```


9. Set of lengths of unique words.
Input: ["cat","dog","lion","cat","tiger","dog"]
Lengths → {3, 4, 5}
Output:
{3, 4, 5}

```python
def unique_word_lengths(words):
    lengths = set()
    for w in set(words):   
        lengths.add(len(w))
    return lengths

words = ["cat","dog","lion","cat","tiger","dog"]
print(unique_word_lengths(words))
```

10. Unique even numbers.
Input: [1,2,2,3,4,4,6,7,8,8]
Output:
{2, 4, 6, 8}

```python
def unique_even_numbers(nums):
    result = set()
    for n in nums:
        if n % 2 == 0:
            result.add(n)
    return result


print(unique_even_numbers([1,2,2,3,4,4,6,7,8,8]))
```

11. Unique characters except vowels.
Input: "datascience"
All unique chars → {d,a,t,a,s,c,i,e,n,c,e}
Remove vowels → {'a','e','i'}
Output:
{'d', 't', 's', 'c', 'n'}

```python
def unique_chars_except_vowels(s):
    vowels = {'a', 'e', 'i', 'o', 'u'}
    result = set()

    for ch in s:
        if ch not in vowels:
            result.add(ch)

    return result

print(unique_chars_except_vowels("datascience"))
```

12. Dictionary of number → cube (1 to 10).
Output:
{1: 1, 2: 8, 3: 27, 4: 64, 5: 125, 6: 216, 7: 343, 8: 512, 9: 729, 10: 1000}

13. Word → length dictionary.
Input: ["apple","kiwi","banana"]
Output:
{'apple': 5, 'kiwi': 4, 'banana': 6}


```python
def word_length_dict(words):
    result = {}
    for w in words:
        result[w] = len(w)
    return result
print(word_length_dict(["apple", "kiwi", "banana"]))
```

14. Filter only even values from dictionary.
Input: {'a':3, 'b':8, 'c':5, 'd':12}
Output:
{'b': 8, 'd': 12}

```python
def filter_even_values(d):
    result = {}
    for key, value in d.items():
        if value % 2 == 0:
            result[key] = value
    return result

data = {'a':3, 'b':8, 'c':5, 'd':12}
print(filter_even_values(data))
```

15. Convert two lists to dictionary.
Input:
keys = ["name","age","city"]
vals = ["Ravi", 25, "Hyderabad"]
Output:
{'name': 'Ravi', 'age': 25, 'city': 'Hyderabad'}

```python
def lists_to_dict(keys, vals):
    result = {}
    for k, v in zip(keys, vals):
        result[k] = v
    return result

keys = ["name", "age", "city"]
vals = ["Ravi", 25, "Hyderabad"]

print(lists_to_dict(keys, vals))
```
