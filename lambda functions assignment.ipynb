Q1. Create a lambda function that converts temperature from Fahrenheit to Celsius.
 Input: temp = 96.8
 Output: 36

```python
to_celsius = lambda f: (f - 32) * 5/9
temp = 96.8
print(int(to_celsius(temp)))
```


Q2. Create a lambda function that finds the area of a circle.
 Input: radius = 5
 Output: 78.54

```python
area_circle = lambda r: 3.1416 * r * r
radius = 5
print(area_circle(radius))
```

Q3. Create a lambda function that converts American dollars to Indian rupees.
 Input: dollar_conversion = 45
 Output: 3950

```python
to_rupees = lambda d: d * 87.78
dollar_conversion = 45
print(int(to_rupees(dollar_conversion)))
```

Q4. Create a lambda function that identifies child and adult (adult ≥ 18).
 Input: age = 10
 Output: Child

```python
check_age = lambda age: "Adult" if age >= 18 else "Child"
age = 10
print(check_age(age))
```

Q6. Write a lambda function that checks if a number is even or odd.

Q7. Create a lambda function that finds the maximum of two numbers.
 Input: 10, 25
 Output: 25

```python
max_two = lambda a, b: a if a > b else b

print(max_two(10, 25))
```

Q10. Write a lambda function to determine if a character is a vowel or consonant.
 Input: 'e'
 Output: Vowel

```python
check_vowel = lambda ch: "Vowel" if ch.lower() in "aeiou" else "Consonant"


print(check_vowel('e')) 
```

Q11. Create a lambda function that checks if a string is empty or not.
 Input: ""
 Output: Empty

```python
is_empty = lambda s: "Empty" if len(s) == 0 else "Not Empty"

print(is_empty("")) 
```


Q12. Write a lambda function that checks if a number lies between 10 and 50 (inclusive).
