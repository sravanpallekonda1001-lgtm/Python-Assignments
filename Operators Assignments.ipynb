Q38. A triangle has sides 3, 4, and 5. Is it a right angled triangle


```python

a = 3
b = 4
c = 5
sides = [a, b, c]
sides.sort() 
x, y, z = sides 
if z**2 == x**2 + y**2:
    print(f"The triangle with sides {a}, {b}, {c} is right-angled.")
else:
    print(f"The triangle with sides {a}, {b}, {c} is NOT right-angled.")
```

Q58. If a is 4. Is the bitwise not of a equal to minus 5
