Write a function that accepts a list of numbers and returns their sum.
Input: [1, 2, 3, 4, 5]
Output: 15


```python
def sum_list(numbers):
    total = 0
    for num in numbers:
        total += num
    return total

print(sum_list([1, 2, 3, 4, 5]))
```

maximum element.
Input: [3, 8, 1, 6, 2]
Output: 8


```python
def max_element(numbers):
    maximum = numbers[0]
    for num in numbers:
        if num > maximum:
            maximum = num
    return maximum

print(max_element([3, 8, 1, 6, 2])) 
```

Write a function that accepts a number and returns whether it is prime or not.
Input: 7
Output: True


```python
def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False

    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

print(is_prime(7))
```

Write a function that accepts a string and returns the number of vowels in it.
Input: "programming"
Output: 3


```python
def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = 0
    for ch in s:
        if ch in vowels:
            count += 1
    return count

print(count_vowels("programming")) 
```

Write a function that accepts two numbers and returns their greatest common divisor (GCD).
Input: a = 12, b = 18
Output: 6


```python
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

print(gcd(12, 18))
```

Write a function that accepts a list of numbers and returns the list after removing duplicates.
Input: [1, 2, 2, 3, 4, 4, 5]
Output: [1, 2, 3, 4, 5]


```python
def remove_duplicates(nums):
    result = []
    for n in nums:
        if n not in result:
            result.append(n)
    return result

print(remove_duplicates([1, 2, 2, 3, 4, 4, 5]))
```

Write a function that accepts a dictionary and returns the key with the maximum value.
Input: {'a': 10, 'b': 20, 'c': 5}
Output: 'b'


```python
def max_key(d):
    return max(d, key=d.get)

print(max_key({'a': 10, 'b': 20, 'c': 5}))
```

Write a function that accepts a number n and returns the factorial of n.
Input: 5
Output: 120


```python
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print(factorial(5))
```


Write a function that accepts a string and returns a dictionary with the frequency of each character.
Input: "apple"
Output: {'a': 1, 'p': 2, 'l': 1, 'e': 1}


```python
def char_frequency(s):
    freq = {}
    for ch in s:
        if ch in freq:
            freq[ch] += 1
        else:
            freq[ch] = 1
    return freq

print(char_frequency("apple"))
```

Input: multiply(2, 3, 4)
Output: 24
Input: multiply(1, 2, 3, 4, 5)

```python
def multiply(*numbers):
    result = 1
    for n in numbers:
        result *= n
    return result
print(multiply(2, 3, 4))        

print(multiply(1, 2, 3, 4, 5)) 
```

Input: print_kwargs(name="Alice")
Output: name: Alice
Input: print_kwargs(name="Bob", age=25, city="Delhi")
