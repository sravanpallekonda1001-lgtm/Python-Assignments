Task : Calculate the following measures for the above data

Mean

Median

Variance

Standard Deviation

IQR

Find the lower fence and upper fence and identify if there are any outliers.


```python
data1 = [16, 14, 22, -11, 16, 23, 25, 19, 12, 27, 21, 14, 60, 85, 13]
import statistics
result=statistics.mean(data1)
result1=statistics.median(data1)
result2=statistics.variance(data1)
result3=statistics.stdev(data1)

print(result)
print(result1)
print(result2)
print(result3)
```

```python
data1.sort()
```

```python
import numpy as np
Q1=np.percentile(data1,25)
Q2=np.percentile(data1,50)
Q3=np.percentile(data1,75)
```

```python
IQR=Q3-Q1
```

```python
lower_fence=Q1-1.5*IQR
upper_fence=Q3+1.5*IQR
```

```python
lower_fence
```

```python
upper_fence
```

```python
outliyer=[x for x in data1 if x<lower_fence and x>upper_fence]
outliyer
```
